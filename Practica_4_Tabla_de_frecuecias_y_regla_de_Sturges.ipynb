## Ejemplo 3: Histogramas de 2 grupos

```python
import random
```

```python
fabrica1=[]
for i in range(100):
  numero_aleatorio= round (random.uniform(50,100),2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
fabrica2=[]
for i in range(100):
  numero_aleatorio2= round (random.uniform(50,100),2)
  fabrica2.append(numero_aleatorio2)
print(fabrica2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(fabrica1)
plt.title("produccion diaria de la fabrica 1")
```

```python
plt.hist(fabrica2)
plt.title("produccion diaria de la fabrica 2")
```

```python
media= mean(fabrica1)
mediana= median(fabrica1)
moda= mode(fabrica1)

print("valores de la fabrica 1")
print("la media es=", media)
print("la mediana es=",mediana)
print("la moda es=",moda)
```

```python
media= mean(fabrica2)
mediana= median(fabrica2)
moda= mode(fabrica2)

print("valores de la fabrica 2")
print("la media es=", media)
print("la mediana es=",mediana)
print("la moda es=",moda)
```

## TABLAS DE FRECUENCIA

```python
# generamos una lista  vacia que representa la edad de un grupo de personas
edades = []
for i in range(100):
       numero_aleatorio = random.randint(18,70)
       edades.append(numero_aleatorio)
print(edades)
```

```python
import pandas as pd
```

```python
df_edades = pd.DataFrame(edades, columns=["edad"])
df_edades
```

```python
#calcular la tabla de ditibucion de frecuencias'
tabla_frecuencia= df_edades["edad"].value_counts().reset_index()
tabla_frecuencia.columns=["edad", "frecuencia"]
tabla_frecuencia
```

```python
#calcular la tabla de ditibucion de frecuencias'
tabla_frecuencia= df_edades["edad"].value_counts().reset_index()
tabla_frecuencia.columns=["edad", "frecuencia"]
tabla_frecuencia= tabla_frecuencia.sort_values(by="edad")
tabla_frecuencia
```

```python
tabla_frecuencia["Frec. Relativa"] = tabla_frecuencia["frecuencia"] / 100
```

```python
tabla_frecuencia
```

```python
tabla_frecuencia["Frec. Porcentaul"] = tabla_frecuencia["frecuencia"] / 100*100
```

```python
tabla_frecuencia
```

```python
#frecuencia acumulada

tabla_frecuencia["Frec.acumulada"] = tabla_frecuencia["frecuencia"].cumsum()
tabla_frecuencia
```

## REGLA DE STURGES


```python
import math
```

```python
print (edades)
```

```python
df_edades_struges= pd.DataFrame({"edades":edades})
df_edades_struges
```

```python
# calcular el n umero de intervalos o clases utilizando la regla de sturges
n= len(edades)
k = int ( 1 + math.log2(n))

print("valor de n=",n,"k o numero de intrvalos o clases =", k)
```

```python
plt.hist(edades)
plt.title("histograma de edades basado en valores estandar matplotlib")
```

```python
tabla_frec_sturges= pd.cut(df_edades_struges["edades"], bins =  k).value_counts().reset_index()
tabla_frec_sturges
```

```python
tabla_frec_sturges.columns= ["clase", " frecuencia"]
tabla_frec_sturges
```

```python
# ordenar la tabla de frecuencias por la clase o el intervalo
tabla_frec_sturges.sort_values(by="clase")
tabla_frec_sturges
```

```python
tabla_frec_sturges["Frec. Relativa"] = tabla_frec_sturges[" frecuencia"] / n
tabla_frec_sturges["Frec. Porcentaul"] = tabla_frec_sturges[" frecuencia"] / n * 100
tabla_frec_sturges["Frec.acumulada"] = tabla_frec_sturges[" frecuencia"].cumsum()
```

```python
tabla_frec_sturges
```

```python
plt.hist(df_edades_struges["edades"], bins=k,alpha=0.5, rwidth=.95)
plt.title("histograma basado en la regla de sturges")
plt.xlabel("edades")
plt.xlabel(" frecuencia")
```
